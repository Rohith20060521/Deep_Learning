# Perceptron(Binary Classification)

```python
import numpy as np
```

# Step 1: Prepare Training Data

```python
x=np.array([1,2,3,4])
y=np.array([0,0,1,1])
x,y
```

# Step 2: Initialize Parameters

```python
w=0.0
b=0.0
alpha=1 # learning rate
epoch=5
print("Initial weight :",w,"bias :",b)
```

# Step 3: Define Activation Function(Step Function)

```python
def step_function(z):
    if z>=0:
        return 1
    else:
        return 0
```

# Training Loop

```python
for epoch in range(epoch):
    print("Epoch :",epoch+1)
    for i in range(len(x)):
        # Compute Weighted sum
        z=w*x[i]+b
        # apply activation function
        y_pred=step_function(z)
        # Compute error
        error=y[i]-y_pred
        # update weights and bias
        w=w+alpha*error*x[i]
        b=b+alpha*error
        print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}, Updated weight: {w}, Updated bias: {b}")
```

# Final Model Parameters

```python
print("Final weight :",w,"bias :",b)
```

# Testing Model

```python
test_value=3.1
z=w*test_value+b
y_pred=step_function(z)
print(f"FOR STUDY HOURS : {test_value}, Predicted : {y_pred}")
```

# Plotting

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c=y,cmap="viridis",label="Data Points")
boundary=-b/w
plt.axvline(x=boundary,color="red",label="Decision Boundary")
plt.title("Perceptron Decision Boundary")
plt.xlabel("Study Hours")
plt.ylabel("Pass/Fail")
```

# Perceptron And Or Implementation

```python
x=np.array([
    [0,0],[0,1],[1,0],[1,1]
])
y=np.array([0,0,0,1])
x,y
```

```python
w=np.zeros(x.shape[1])
b=0
alpha=0.1
epoch=10
print("Initial weights :",w,"bias :",b)
```

```python
def step_function(z):
    if z>=0:
        return 1
    else:
        return 0
```

```python
for i in range(epoch):
    print("Epoch :",i+1)
    for j in range(len(x)):
        z=np.dot(w,x[j])+b
        y_pred=step_function(z)
        error=y[j]-y_pred
        w=w+alpha*error*x[j]
        b=b+alpha*error
        print(f"Input: {x[j]}, Predicted: {y_pred}, Actual: {y[j]}, Error: {error}, Updated weights: {w}, Updated bias: {b}")
```

```python
print("Final weight :",w,"bias :",b)
```

```python
for i in range(len(x)):
    z=np.dot(w,x[i])+b
    y_pred=step_function(z)
    print(f"FOR INPUT : {x[i]}, Predicted : {y_pred}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c=y,cmap="viridis",label="Data Points")
x=np.linspace(0,1,100)
x2=-(w[0]*x+b)/w[0]
plt.plot(x,x2,color="red",label="Decision Boundary")
plt.title("Perceptron Decision Boundary")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
```

# Email Classification

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([
    [1, 1],   # E1
    [1, 0],   # E2
    [0, 1],   # E3
    [0, 0]    # E4
])
y = np.array([1, 1, 1, 0])
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 20
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = 1 if linear_output >= 0 else 0 
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error
        print(f"Epoch: {epoch+1}, Sample: {i}, Target: {y[i]}, Predicted: {prediction}, Error: {error}, Updated Weights: {weights}, Updated Bias: {bias}")

print("Final Weights:", weights)
print("Final Bias:", bias)
x1_vals = np.linspace(-0.5, 1.5, 100)
if weights[1] != 0:
    x2_vals = (-weights[0] * x1_vals - bias) / weights[1]
else:
    x2_vals = np.zeros_like(x1_vals)
plt.scatter(X[y==1][:,0], X[y==1][:,1])
plt.scatter(X[y==0][:,0], X[y==0][:,1])
plt.plot(x1_vals, x2_vals)
plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.title("Perceptron Decision Boundary (Spam Detection)")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.show()
```
